# Fix parameters:
## Number of imputs = 2, slope and intersect
## w_0 = (0,0)
## w_true = (2,2)
## batch_size = Number of imputs/10  // For mini-batch experiments

```python
num_inputs = 2 # Including the dim of bias
true_w = [2, 2]
w = np.array([[0],[0]])
#w = np.random.normal(scale=1, size=(num_inputs, 1))
print('Initial value of the parameter \n',w)
```

## alpha = 0.1

```python
num_examples = 100
a = OptiObj(num_inputs, num_examples, true_w)
# Fix Step size
## alpha = 0.4
a.init_params(w)
a.train(batch_size=1, lr=0.1, epochs=10, period=9)
a.init_params(w)
a.train(batch_size=num_examples, lr=0.1, epochs=50, period=num_examples,  app = 'GD')
a.init_params(w)
a.train(batch_size=10, lr=0.1, epochs=15, period=120,  app = 'BG')
# Fix Step size
## alpha = 0.04
```

## Alpaha = 0.01

```python
num_examples = 100
a = OptiObj(num_inputs, num_examples, true_w)
# Fix Step size
## alpha = 0.4
a.init_params(w)
a.train(batch_size=1, lr=0.01, epochs=15, period=9)
a.init_params(w)
a.train(batch_size=num_examples, lr=0.01, epochs=400, period=num_examples,  app = 'GD')
a.init_params(w)
a.train(batch_size=10, lr=0.01, epochs=120, period=120,  app = 'BG')
```
